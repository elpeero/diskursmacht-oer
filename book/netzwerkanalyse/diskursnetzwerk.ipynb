## Distant Reading: die Darstellung von diskursemantischem Wissen in Netzwerken

Um den         Diskurs √ºberblickhaft aus einer Makroperspektive zu betrachten m√ºssen die konstitutiven Texte aus der Distanz betrachtet werden, im sogenannten Distant Reading. F√ºr den Zweck das semantische Wissen zu erfassen werden die explorativ annotierten Frames in einen Netzwerkgraphen √ºberf√ºhrt. Da aber auch nachvollzogen werden soll, falls zu einem bestimmten Diskurs tendieren, werden auch die Schl√ºsselwortkategorien abgebildet.

`````python
import gravis as gv
import pandas as pd
import networkx as nx
from collections import Counter

# Set color scheme for graph
color_dict = {
    'Bildung': "#d5a700",     
    'Schule': '#d5a700',
    'Kita': '#d5a700',
    'Ausbildung': "#d5a700",
    'Berufst√§tigkeit': '#882255',  
    'Arbeitsmarkt': '#882255',
    'Migration': '#0072b2',   
    'Integration': '#0072b2',
    'Deutsch': '#0072b2',
    'Sprache': '#0072b2',
    'Sprachkurs': '#0072b2',
    'Gefl√ºchtete': '#0072b2',
}

border_color_dict = {
    '#d5a700': "#ac8600",
    '#882255': '#5e1839',
    '#0072b2': '#005280'
}


date_dict = {
    '01': 'Janur',
    '02': 'Februar',
    '03': 'M√§rz',
    '04': 'April',
    '05': 'May',
    '06': 'Juni',
    '07': 'Juli',
    '08': 'August',
    '09': 'September',
    '10': 'Oktober',
    '11': 'November',
    '12': 'Dezember'
}


central_ls = [
 'Schwierigkeit',
 'Bildung_Lehre',
 'Umfassen',
 'Unterst√ºtzen',
 'Voraussetzung_Szenario',
 ]

frames_df = pd.read_csv('network_data.txt', sep='\t')

key_ls = []
val_ls = []
disc_ls = []
desc_ls = []
num_ls = []

num_dict = {}
frame_dict = {}

for index, row in frames_df.iterrows():
    value = row['frame']
    lu = row['instantiation']
    date = row['datum']

for index, row in frames_df.iterrows():
    keys = row['kw_type'].split('; ')
    value = row['frame']
    disc = row['diskurs']

    for key in keys:
        key_ls.append(key)
        val_ls.append(value)
        disc_ls.append(disc)

kw_dict = {
    'id': val_ls,
    'keyword': key_ls,
    'frame':val_ls,
    'group': disc_ls,
}
size_dict = dict(Counter(key_ls+val_ls))

data = pd.DataFrame(kw_dict)


G = nx.Graph()

for id, size in size_dict.items():
    if id in key_ls:
        col = color_dict[id]
        G.add_node(id, size=size, color=col, label_color='black', border_color=border_color_dict[col], border_size=2, label=id, label_size=10)

    else:
        G.add_node(id, size=size, shape='rectangle', opacity=1)

for _, row in data.iterrows():
    source = row['keyword']
    target = row['frame']
    node_color = G.nodes[source]['color']
    # Check if edge already exists
    if G.has_edge(source, target):
        # Increment the weight
        G[source][target]['weight'] = G[source][target].get('weight', 0) + 1
    else:
        G.add_edge(source, target, color=node_color, opacity=1, size=2, weight=1)
for node in G.nodes:
    # Only process rectangular nodes (frames)
    if G.nodes[node].get('shape') == 'rectangle':
        num_ls.append(G.nodes[node]['size'])

numbers = set(sorted(num_ls))
for i, j in enumerate(numbers):
    n = i+1
    num_dict[j] = n
    
for node in G.nodes:
    # Only process rectangular nodes (frames)
    if G.nodes[node].get('shape') == 'rectangle':
        # Find all edges connected to this node
        connected_edges = G.edges(node, data=True)

        # Count the frequency of each edge color
        color_counter = Counter()
        for _, _, edge_data in connected_edges:
            color = edge_data.get('color')
            weight = edge_data.get('weight')
            if color:
                color_counter[color] += 1*weight

        # If no colors found, skip
        if not color_counter:
            continue

        # Find dominant colors
        max_freq = max(color_counter.values())
        dominant_colors = [c for c, count in color_counter.items() if count == max_freq]

        if len(dominant_colors) == 1:
            col = dominant_colors[0]
            G.nodes[node]['color'] = col
            G.nodes[node]['label_color'] = border_color_dict[col]
            G.nodes[node]['border_color'] = border_color_dict[col]
        else:
            G.nodes[node]['color'] = "#282828"
            G.nodes[node]['label_color'] = "#282828"
            G.nodes[node]['border_color'] = "#000000"

        G.nodes[node]['border_size'] = 1

        G.nodes[node]['label'] = str(node).upper()
        G.nodes[node]['label_size'] = 5 + (num_dict[G.nodes[node]['size']]/2)


node_cols = {node: G.nodes[node].get('color') for node in G.nodes}


G.graph['background_color'] = "whitesmoke"


fig = gv.d3(G,
    graph_height=1000,
    details_height=200,
    show_details=False,                      
    show_details_toggle_button=False,      
    node_label_data_source='label',
    node_drag_fix=True,        

    node_hover_neighborhood=True,
    zoom_factor=1.2,

    use_node_size_normalization=True,
    node_size_normalization_max=30,

    use_edge_size_normalization=True,
    edge_size_data_source='weight',
    edge_curvature=0.3,
    edge_size_normalization_max=2,

    use_many_body_force=True,
    many_body_force_strength=-150,
    many_body_force_theta=1.2,

    use_links_force=True,
    links_force_distance=80,
    links_force_strength=0.15,

    use_x_positioning_force=True,
    x_positioning_force_strength=0.1,
    use_y_positioning_force=True,
    y_positioning_force_strength=0.05,

    use_collision_force=False,
    collision_force_radius=5.0,
    collision_force_strength=0.5,

    use_centering_force=True,

    layout_algorithm_active=True,
      )

fig
`````
